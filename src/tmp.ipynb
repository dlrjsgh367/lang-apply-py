```python
import re

# +로 양쪽이 감싸진 텍스트를 탐지하는 정규식
plus_wrapped_pattern = re.compile(r"\+(.*?)\+")

# 테스트 데이터
texts = [
    'alertContent: "금칙어 \'" + keyword + "\'가 포함되어 노무 상담 등록이 불가능합니다.",',
    'This is a test string without plus signs.',
    'Another +example+ of wrapped text.',
    'No wrapping here + only single.',
]

# 매칭 탐지
for text in texts:
    match = plus_wrapped_pattern.search(text)
    if match:
        print(f"Match in '{text}': {match.group(1)}")
```
